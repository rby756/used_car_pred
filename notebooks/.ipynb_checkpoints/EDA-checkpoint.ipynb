```python
# Function to check for unique values in every categorical column
def check_unique_values_in_categorical_columns(df):
    categorical_columns = df.select_dtypes(include=['object', 'category']).columns
    unique_values_dict = {}
    for column in categorical_columns:
        unique_values = df[column].unique().tolist()
        unique_values_dict[column] = unique_values
    return unique_values_dict
```

```python


# Function to check if values in a particular column are in the list of acceptable values
def check_column_values(df, column, values):
    invalid_values = df[df[column].isin(values)]
    if not invalid_values.empty:
#         print(f"Rows with invalid values in column '{column}':\n", invalid_values)
        return invalid_values
    else:
#         print(f"All values in column '{column}' are within the acceptable values.")
        return None
        
    
```

```python
# Call the function to check the DataFrame
unique_category=check_unique_values_in_categorical_columns(data)
```

```python
#dropped a row which have wrong info in multiple columns
data=data[data['registration_year']!='Third Party insurance']
```

```python
# Function to update registration values
def update_registration_year(value):
    if '-' in value:
        # Extract the year part after the hyphen
        year_part = value.split('-')[-1]
        # Ensure the year part is two digits and convert to 20xx format
        year = f"20{year_part.zfill(2)}"
    else:
        # Assume the value is already a year
        year = f"20{value[-2:]}"
    return year

# Apply the function to the 'Registration' column
data['registration_year'] = data['registration_year'].apply(update_registration_year)
```

```python
#Outlier removal 
min_threshold,max_threshold = data['seats'].quantile([0.001,0.999])
min_threshold1, max_threshold1 = data['mileage(kmpl)'].quantile([0.001,0.999])
min_threshold2, max_threshold2 = data['kms_driven'].quantile([0.001,0.999])
min_threshold3, max_threshold3 = data['engine(cc)'].quantile([0.001,0.999])
min_threshold4, max_threshold4 = data['torque(Nm)'].quantile([0.001,0.999])
min_threshold5, max_threshold5 = data['price(in lakhs)'].quantile([0.001,0.999])
```

```python
data = data[(data['seats']<max_threshold)&(data['seats']>=min_threshold)]
data = data[(data['mileage(kmpl)']<max_threshold1)&(data['mileage(kmpl)']>=min_threshold1)]
data = data[(data['kms_driven']<max_threshold2)&(data['kms_driven']>=min_threshold2)]
data = data[(data['engine(cc)']<max_threshold3)&(data['engine(cc)']>=min_threshold3)]
data = data[(data['torque(Nm)']<max_threshold4)&(data['torque(Nm)']>=min_threshold4)]
data = data[(data['price(in lakhs)']<max_threshold5)&(data['price(in lakhs)']>=min_threshold5)]
data.reset_index(drop=True,inplace=True)
```

```python
data.drop(['index'],inplace=True,axis=1)
```

```python
# Function to extract 2nd and 3rd words from a string
def extract_2nd_3rd_words(carname):
    words = carname.split()
    if len(words) >= 3:
        return f"{words[1]} {words[2]}"
    return ""

# Apply the function to the 'carname' column
data['short_carname'] = data['car_name'].apply(extract_2nd_3rd_words)
```
